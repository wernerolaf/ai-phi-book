### Data

Random Variable is the source of data which we observe. It generates outcomes with given probabilities determined by the probability distribution. It is unknown as it is outside of reality. Our reality is simply one of many possible outcomes. Our observed data was created by the unknown statistical process. In that regard random variable and probability distribution are both outside reality as transcendental abstractions and are more real than reality as random variable is first in causal chain. Existence of dice precede existence of rolled outcomes.

- **Probability Distribution**  
  A function that describes how probability is assigned to the values of a random variable $X$. It is how weighted our dice is.

  **Discrete case, Probability Mass Function (PMF):**
  $$
  P(X = x) = p(x)
  $$
  $$
  \sum_x p(x) = 1
  $$

  **Continuous case, Probability Distribution Function (PDF):**
  $$
  f_X(x) \ge 0
  $$
  $$
  \int_{-\infty}^{\infty} f_X(x)\,dx = 1
  $$

  **Cumulative Distribution Function (CDF):**
  $$
  F_X(x) = P(X \le x)
  $$

- **Sample/Data/Observation**  
  A finite set of observed values drawn from a population or distribution. These are rolled outcomes of the dice.
  A sample of size $n$ is written as:
  $$
  \{x_1, x_2, \dots, x_n\}
  $$
  Unfortunately sometimes this vector is also denoted as X which can be confusing.

- **Random Variable**  
  A numerical function that maps outcomes of a random experiment to real numbers.  
  It is commonly denoted by: $X$. It is our dice.
  To avoid issues related to measure theory we also assume that this function is measurable


In normal world data and information are used interchangeably. In case of statistic this is serious mistake.

- **Entropy**  
  A measure of the uncertainty or randomness of a random variable $X$.  
  Higher entropy means more unpredictability.

  **Discrete entropy:**
  $$
  H(X) = -\sum_{x} P(X=x)\log P(X=x)
  $$

  **Continuous entropy (Differential Entropy):**
  $$
  H(X) = -\int_{-\infty}^{\infty} f_X(x)\log f_X(x)\,dx
  $$

  **Properties:**
  - $ H(X) \ge 0 $
  - Maximum entropy occurs when all outcomes are equally likely

  **Units:**
  - Base 2 logarithm → bits
  - Base $e$ logarithm → nats


- **Joint Entropy**  
  Measures the uncertainty of two random variables $X$ and $Y$ together.

  $$
  H(X, Y) = -\sum_{x,y} P(X=x, Y=y)\log P(X=x, Y=y)
  $$

- **Conditional Entropy**  
  Measures the remaining uncertainty of $X$ given that $Y$ is known.

  $$
  H(X \mid Y) = -\sum_{x,y} P(X=x, Y=y)\log P(X=x \mid Y=y)
  $$

  **Relationship:**
  $$
  H(X, Y) = H(X \mid Y) + H(Y)
  $$

- **Mutual Information**  
  Measures how much information one random variable contains about another.  
  Widely used in feature selection, representation learning, and many other use cases.

  $$
  I(X; Y) = \sum_{x,y} P(X=x, Y=y)\log \frac{P(X=x, Y=y)}{P(X=x)P(Y=y)}
  $$

  **Equivalent forms:**
  $$
  I(X; Y) = H(X) - H(X \mid Y)
  $$
  $$
  I(X; Y) = H(X) + H(Y) - H(X, Y)
  $$

  **Properties:**
  - $I(X; Y) \ge 0$
  - $I(X; Y) = 0$ if and only if $X$ and $Y$ are independent

- **Cross-Entropy**  
  Measures the average number of bits needed to encode data from a true distribution $P$ using a model distribution $Q$.  
  This is the standard loss function for classification models.

  $$
  H(P, Q) = -\sum_x P(x)\log Q(x)
  $$

- **Kullback–Leibler (KL) Divergence**  
  Measures how much one probability distribution $Q$ diverges from a true distribution $P$.  
  It is **not symmetric** and **not a true distance metric**.

  $$
  D_{\mathrm{KL}}(P \| Q) = \sum_x P(x)\log \frac{P(x)}{Q(x)}
  $$

- **Relationship: Cross-Entropy vs KL Divergence**

  $$
  H(P, Q) = H(P) + D_{\mathrm{KL}}(P \| Q)
  $$

  **Implication for ML:**
  - Minimizing cross-entropy is equivalent to minimizing KL divergence
  - Since $H(P)$ is fixed, learning focuses on reducing $D_{\mathrm{KL}}(P \| Q)$



To summarize information is difference between starting entropy of the Random Variable and its end entropy under assumption that we observed different variable. So information is contextual property of data.

There is also dualistic interpretation of entropy. It is also average number of bits needed to encode certain outcomes. Cross Entropy and Kullback-Leibler Divergence measure number of bits used to encode the outcomes when we are using our model so fake/estimated Random Variable and we want to minimize unnecessery bits. The most succint, (but not less!) description of reality is also the most accurate model.

So Random Variables are not variables, but functions. Data is not information, but can contain information. Entropy is bad so we want to get rid of it.

# Model

So what is the essence of the modelling?

$min_{\theta \in \Theta} -\frac{1}{N} \sum_{1}^{N} \log Q_{\theta}(x_i)$

Big $\Theta$ is parameter space of our model $Q$, also called family of random variables. For example for neural network it would be set of all possible weights combinations. $\theta$ is some concrete parametrization of this Random Variable.

$-\frac{1}{N} \sum_{1}^{N} \log Q_{\theta}(x_i)$ is empirical version of the cross entropy. In statistics formulas often have two versions: theoretical and empirical. Only empirical formulas can be evaluated, which means actual computer can calculate some value using observed data. So they are estimates. Theoretical formulas can even contain latent random variables. Latent random variables are variables for which we never even observe data. Dices completely behind the scenes. But can be used in model creation and analysis.

So in the end we want to find the best, most plausible model using some optimization process.

Final pipeline is this:

1. You want to make accurate predictions
2. To make good predictions you need to reduce entropy
3. To reduce entropy you need information
4. To get information you need data
5. To extract information from data you need a model
6. To train a model you need to perform optimization
7. Spend unknown amount of time dealing with ridiculous issues

After all trails and tribulations we get model.
Model is approximation of the true distribution. As we do not know truth we need to estimate it based on the observed outcomes. What we want to do is to recreate dice based on the observed outcomes. Predictions are simply imperfectly recreated realities.
