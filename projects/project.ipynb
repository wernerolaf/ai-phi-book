# Linear Regression Toy Example

A tiny project notebook that fits a line to synthetic data.

We generate points from $y = 2x + 1$ with small noise and recover the slope using least squares.

```
import numpy as np
np.random.seed(1)
x = np.linspace(0, 1, 6)
noise = np.random.normal(scale=0.02, size=x.shape)
y = 2 * x + 1 + noise
A = np.vstack([x, np.ones_like(x)]).T
coeffs, *_ = np.linalg.lstsq(A, y, rcond=None)
coeffs
```

The estimated slope and intercept should be close to $(2, 1)$ on each run.
